```python
# 스크롤을 끝까지 내리는 함수 
def scroll_to_bottom_of_element(element):
    '''빠르게 스크롤을 요소의 끝까지 내리는 함수'''
    while True:
        # 현재 스크롤 위치
        current_scroll_position = driver.execute_script("return arguments[0].scrollTop", element)
        
        # 큰 단위로 스크롤
        driver.execute_script("arguments[0].scrollTop += 1000", element)
        
        # 잠깐 대기
        time.sleep(0.5)
        
        # 새로운 스크롤 위치 확인
        new_scroll_position = driver.execute_script("return arguments[0].scrollTop", element)
        if new_scroll_position == current_scroll_position:
            break
```

```python
# 무한 스크롤 함수 정의
def infinite_scroll(driver):
    body = driver.find_element(By.CSS_SELECTOR, 'body')
    body.send_keys(Keys.END)

# 모델명 찾기
def find_model_name(driver, i):
    try:
        model_name = WebDriverWait(driver, 1).until(
            EC.presence_of_element_located((By.CSS_SELECTOR, f"#__layout > div > div.layout__main.container.search.lg > div.content > div > div:nth-child(3) > div > div.search_result_list > div:nth-child({i}) > div > a > div.product_info_area > div.title > div > p.translated_name"))
        )
        return model_name.text
    except :
        return "N/A"
```

```python
import re

# 가격정보는 숫자로 넣어야되니까.... 바꿔..................야됨..........
def convert_to_int(data):
    # 거래된 가격을 int형으로
    data[1] = int(data[1])

    # 출시가격을 int형으로
    # 4번 인덱스의 값에서 숫자를 추출하여 int로 변환
    price_match = re.search(r'약?\s*([\d,]+)원', data[3])
    if price_match:
        # 쉼표를 제거하고 숫자를 변환
        data[3] = int(price_match.group(1).replace(',', ''))
    else:
        data[3] = 0  # 숫자를 찾지 못했을 경우 0으로 설정
    
    return data
for data in deal_data :
    try :
        data = convert_to_int(data)
    except:
        continue
        
for data in buy_list :
    try :
        data[1] = int(data[1].replace(',', '').replace('원', ''))
    except:
        continue

for data in sell_list :
    try :
        data[1] = int(data[1].replace(',', '').replace('원', ''))
    except:
        continue
    
```

```python
import csv

# DEAL_LIST.csv, RELEASE_PRICE.csv 파일 저장

# DEAL_LIST.csv 파일로 저장할 데이터 필터링
# model, price, date
deal_list_filtered = []
for deal in deal_data:
    # deal의 길이를 확인하고 필요한 데이터만 추가
    if len(deal) >= 3:
        deal_list_filtered.append([deal[0], deal[1], deal[2]])


# DEAL_LIST.csv 파일로 이어쓰기
deal_list_csv_filename = 'X_DEAL_LIST.csv'
deal_list_columns = ['model', 'price', 'date']
file_exists = os.path.isfile(deal_list_csv_filename)

with open(deal_list_csv_filename, mode='a', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    
    # 파일이 처음 생성될 때 컬럼 이름 작성
    if not file_exists:
        writer.writerow(deal_list_columns)
    
    # 데이터 추가
    writer.writerows(deal_list_filtered)

```
